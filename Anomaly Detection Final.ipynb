# Credit Card Fraud Detection
This notebook will guide you through the process of building a machine learning model to detect fraudulent transactions. We will be using the 2013 European credit card fraud dataset from Kaggle. The steps we will follow are:

1. **Exploratory Data Analysis (EDA)**: We will start by loading the dataset and performing an initial analysis to understand the data. This will include checking the data types, missing values, summary statistics, and visualizing the distributions of variables.

2. **Data Preprocessing**: The dataset is highly imbalanced, so we will apply techniques to balance it. This could include oversampling the minority class (fraudulent transactions), undersampling the majority class (non-fraudulent transactions), or using a combination of both.

3. **Feature Selection**: We will identify the most relevant or correlated variables with the 'Class' variable, which indicates whether a transaction is fraudulent or not. This can be done using correlation analysis, chi-square test, or any other suitable feature selection method.

4. **Model Training**: We will train four different models - Local Outlier Factor, K-means clustering, Support Vector Machines (SVM), and XGBoost - using the selected features. We will use cross-validation to ensure the robustness of the models.

5. **Hyperparameter Optimization**: We will optimize the hyperparameters used to improve our scores and lower our models' mistakes with its predictions.

6. **Model Evaluation**: We will evaluate the performance of each model using precision, F1-score, AUC-ROC, and recall metrics. Also, we will create confusion matrices to evaluate the models based on True Positives (TP), True Negatives (TN), False Positives (FP), and False Negatives (FN).

7. **Model Comparison**: We will compare the performance of the four models and select the best performing one. If other models compensate for the insufficiencies of others, we will suggest a hybrid model.

Each step will be well-documented with markdown cells explaining the process and findings. We will also include necessary visualizations for better understanding.

```python
from sklearn.model_selection import train_test_split

# Separate input features and target
y = df['Class']
X = df[['V4','V11','V2','V19','V20','V14','V12','V10','V16','V17']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import LocalOutlierFactor

# Initialize the model with novelty = False for Classification report 
lof_false = LocalOutlierFactor(n_neighbors=5, algorithm='auto', 
                                              leaf_size=30, metric='minkowski',
                                              p=2, metric_params=None,contamination=0.1)

# Fit the model
lof_false.fit(X_train)
```

```python
from sklearn.neighbors import LocalOutlierFactor

# Initialize the model novelty = True for confusion matrix
lof = LocalOutlierFactor(n_neighbors=5, algorithm='auto', 
                                              leaf_size=30, metric='minkowski', novelty=True,
                                              p=2, metric_params=None,contamination=0.1)

# Fit the model
lof.fit(X_train)
```

```python
from sklearn.cluster import KMeans

# Initialize the model
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)

# Fit the model
kmeans.fit(X_train)
```

```python
from sklearn import svm

# Initialize the model
clf = svm.SVC(probability=True,kernel='rbf', degree=3, gamma=0.1,max_iter=-1, random_state=42)

# Fit the model
clf.fit(X_train, y_train)
```

```python
!pip install xgboost
```

```python
import xgboost as xgb

# Initialize the model
xg_clf = xgb.XGBClassifier(learning_rate = 0.2, n_estimators= 100, objective='binary:logistic',
                           max_depth= 5, min_child_weight= 3, max_delta_step= 1, subsample= 0.8)

# Fit the model
xg_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Predict the labels for the test set
y_pred_lof = lof_false.fit_predict(X_test)
y_pred_kmeans = kmeans.predict(X_test)
y_pred_svm = clf.predict(X_test)
y_pred_xgb = xg_clf.predict(X_test)

# Map the LOF predictions from -1 and 1 to 0 and 1 to match your target variable.
y_pred_lof = (y_pred_lof == 1).astype(int)

# Print the classification report for each model
print('Classification report for LOF:')
print(classification_report(y_test, y_pred_lof))

print('Classification report for K-means:')
print(classification_report(y_test, y_pred_kmeans))

print('Classification report for SVM:')
print(classification_report(y_test, y_pred_svm))

print('Classification report for XGBoost:')
print(classification_report(y_test, y_pred_xgb))
```

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix
import matplotlib.pyplot as plt

# List of all models and their names for ROC curve
models_roc = [(lof_false, 'LOF'), (kmeans, 'K-means'), (clf, 'SVM'), (xg_clf, 'XGBoost')]
fig, ax = plt.subplots()

# Loop through all models
for model, name in models_roc:
    # Compute predicted probabilities
    if name == 'LOF' or name == 'K-means':
        y_score = model.fit_predict(X_test)
    else:
        y_score = model.predict_proba(X_test)[:,1]
    
    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    ax.plot(fpr, tpr, label='%s ROC curve (area = %0.2f)' % (name, roc_auc))

# Plot random guess line
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

# Set plot labels and legends
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

# Show plot
plt.show()

# List of all models and their names for confusion matrix
models_cm = [(lof, 'LOF'), (kmeans, 'K-means'), (clf, 'SVM'), (xg_clf, 'XGBoost')]

import seaborn as sns

# Print confusion matrices
for model, name in models_cm:
    # Predict the labels for the test set
    y_pred = model.predict(X_test)
    
    # Map the predicted labels to match with the 'Class' variable encoding only for LOF model
    if name == 'LOF':
        y_pred = np.where(y_pred == -1, 1, 0)

    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Create a DataFrame from the confusion matrix
    cm_df = pd.DataFrame(cm, columns=['Predicted Negative', 'Predicted Positive'], 
                         index=['Actual Negative', 'Actual Positive'])
    
    # Create a heatmap from the DataFrame
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    
    # Set the plot title
    plt.title('Confusion Matrix for %s' % name)
    
    # Show the plot
    plt.show()
```

# Random Undersampling

In the previous steps, we used SMOTE for handling the imbalanced dataset. Now, we will try another technique known as Random Undersampling, which works by eliminating instances from the majority class to match the minority class. Let's implement this and see how it affects our models.

```python
from imblearn.under_sampling import RandomUnderSampler

# define undersample strategy
undersample = RandomUnderSampler(sampling_strategy='majority')

# fit and apply the transform
X_under, y_under = undersample.fit_resample(X, y)

# summarize class distribution
print('After undersampling: class=%d, n=%d' % (0, sum(y_under == 0)))
print('After undersampling: class=%d, n=%d' % (1, sum(y_under == 1)))
```

```python
from sklearn.model_selection import train_test_split

# Split the undersampled data into training and testing sets
X_train_under, X_test_under, y_train_under, y_test_under = train_test_split(X_under, y_under, test_size=0.25, random_state=42)

```

The output shows that we now have an equal number of instances for both classes. We can now proceed to train our models using this undersampled data and compare the results with the models trained using SMOTE.

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix
import matplotlib.pyplot as plt

# List of all models and their names for ROC curve
models_under_roc = [(lof_under_false, 'LOF'), (kmeans_under, 'K-means'), (clf_under, 'SVM'), (xg_clf_under, 'XGBoost')]
fig, ax = plt.subplots()

# Loop through all models
for model, name in models_under_roc:
    # Fit the model
    model.fit(X_train_under, y_train_under)
    
    # Compute predicted probabilities
    if name == 'LOF' or name == 'K-means':
        y_under_score = model.fit_predict(X_test_under)
    else:
        y_under_score = model.predict_proba(X_test_under)[:,1]
    
    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(y_test_under, y_under_score)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    ax.plot(fpr, tpr, label='%s ROC curve (area = %0.2f)' % (name, roc_auc))

# Plot random guess line
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

# Set plot labels and legends
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

# Show plot
plt.show()


# List of all models and their names for confusion matrix
models_cm = [(lof_under_true, 'LOF'), (kmeans_under, 'K-means'), (clf_under, 'SVM'), (xg_clf_under, 'XGBoost')]

import seaborn as sns

# Print confusion matrices
for model, name in models_cm:
    # Predict the labels for the test set
    y_pred_under = model.predict(X_test_under)
    
    # Map the predicted labels to match with the 'Class' variable encoding only for LOF model
    if name == 'LOF':
        y_pred_under = np.where(y_pred_under == -1, 1, 0)

    # Compute confusion matrix
    cm = confusion_matrix(y_test_under, y_pred_under)
    
    # Create a DataFrame from the confusion matrix
    cm_df = pd.DataFrame(cm, columns=['Predicted Negative', 'Predicted Positive'], 
                         index=['Actual Negative', 'Actual Positive'])
    
    # Create a heatmap from the DataFrame
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    
    # Set the plot title
    plt.title('Confusion Matrix for %s' % name)
    
    # Show the plot
    plt.show()

```

# Model Interpretation


## Local Outlier Factor (LOF)
The LOF model performed poorly on both the undersampled and SMOTE data. The precision for detecting fraudulent transactions was 0.54 for the undersampled data and 0.00 for the SMOTE data. This indicates that the model is not good at correctly identifying positive instances (fraudulent transactions) among all the instances it labeled as positive. The recall was high (0.91 for SMOTE and 0.11 for undersampled), which means the model was able to identify a high proportion of actual positive instances. However, the low precision means that many of the instances it identified as positive were actually negative. The f1-score, which is a measure of the model's accuracy, was also low for both datasets.

The confusion matrix for the LOF model on the SMOTE data shows a high number of false positives (8750), indicating that the model incorrectly labeled many non-fraudulent transactions as fraudulent. On the undersampled data, the model had a high number of false negatives (492), meaning it failed to identify many fraudulent transactions.

## K-means
The K-means model performed better on the undersampled data than on the SMOTE data. The precision, recall, and f1-score were all higher for the undersampled data. However, the model still had a relatively high number of false positives (27139) on the SMOTE data, indicating that it incorrectly labeled many non-fraudulent transactions as fraudulent.

## Support Vector Machine (SVM)
The SVM model performed well on both the undersampled and SMOTE data. It had high precision, recall, and f1-scores for both datasets, indicating that it was good at correctly identifying fraudulent transactions and had a low rate of false positives and false negatives. The confusion matrices for the SVM model show a low number of false positives and false negatives, further indicating the model's good performance.

## XGBoost
The XGBoost model performed exceptionally well on both the undersampled and SMOTE data. It had perfect precision, recall, and f1-scores for the undersampled data, and near-perfect scores for the SMOTE data. This indicates that the model was excellent at correctly identifying fraudulent transactions and had a very low rate of false positives and false negatives. The confusion matrices for the XGBoost model show a very low number of false positives and false negatives, further indicating the model's excellent performance.

In conclusion, the SVM and XGBoost models performed the best on both the undersampled and SMOTE data. The LOF model performed the worst, and the K-means model performed moderately well on the undersampled data but poorly on the SMOTE data. These results suggest that SVM and XGBoost are the most suitable models for this task, while LOF is the least suitable.
